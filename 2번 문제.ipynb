### Keras

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
data=pd.read_csv('trees.csv')
```

```python
xdata=data.iloc[:,:-1]
ydata=data.iloc[:,[-1]]
```

```python
IO=Dense(units=1,input_dim=2)
model=Sequential([IO])
model.compile(loss='mean_squared_error',optimizer=Adam(0.1))
h=model.fit(xdata,ydata,epochs=1000)
```

```python
plt.plot(h.history['loss'])
plt.show()
```

```python
wb=IO.get_weights()
wb
```

```python
# numpy 행렬로 예측하는 법
model.predict(np.array([[8.8,63],[10.5,72]]))
```

```python
plt.plot(ydata)
plt.plot(model.predict(xdata),'r--')
plt.show()
```
